```python
import numpy as np
from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import chi2
from scipy.stats import kstat
import matplotlib.pyplot as plt
from statsmodels.distributions.edgeworth import ExpandedNormal
```

```python
#compute cumulants of an arbitrary distribution 
data = uniform.rvs(size=1000)

#https://tedboy.github.io/statsmodels_doc/_modules/statsmodels/distributions/edgeworth.html
kappa = []
for i in range(1,5):
    kappa.append(kstat(data,n=i))


    
edge_expansion = ExpandedNormal(kappa)

x = np.linspace(0,1,20)
plt.plot(x,edge_expansion.pdf(x),label="edge")
plt.plot(x,uniform.pdf(x),label="unif")
plt.legend()
plt.show()
```

```python
#compute cumulants of an arbitrary distribution 
df = 4
data = chi2(df).rvs(size=1000)

#https://tedboy.github.io/statsmodels_doc/_modules/statsmodels/distributions/edgeworth.html
kappa = []
for i in range(1,5):
    kappa.append(kstat(data,n=i))

print(kappa)
    
edge_expansion = ExpandedNormal(kappa)

x = np.linspace(0,20,100)
plt.plot(x,edge_expansion.pdf(x),label="edge")
plt.plot(x,chi2(df).pdf(x),label="chi2")
plt.legend()
plt.show()
```

```python
#compute cumulants of an arbitrary distribution 
data = norm.rvs(size=1000)

#https://tedboy.github.io/statsmodels_doc/_modules/statsmodels/distributions/edgeworth.html
kappa = []
for i in range(1,5):
    kappa.append(kstat(data,n=i))

print(kappa)
    
edge_expansion = ExpandedNormal(kappa)

x = np.linspace(-1,1,20)
plt.plot(x,edge_expansion.pdf(x),label="edge")
plt.plot(x,norm.pdf(x),label="norm")
plt.legend()
plt.show()
```
